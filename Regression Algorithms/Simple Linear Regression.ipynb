**Aim:** Trying to explain the salary level by the years of experience <code> Salary = b0 + b1*Experience</code>.

#### ‚òëÔ∏è Importing Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### üëâData Preprocessing

```python
df = pd.read_csv('C:/Users/HP Elitebook/OneDrive/Bureau/HOMEWORKS/ML_BOOTCAMP/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv')
df.head(10)
```

```python
df.info()
```

```python
x = df.iloc[:,0].values
x = x[:,np.newaxis]
y = df.iloc[:,1].values
y = y[:,np.newaxis]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### üëâ Training SLRM on the training sets

#### ‚òëÔ∏è Fitting the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

#### ‚òëÔ∏è Predictions

```python
y_pred = regressor.predict(x_test)
```

#### ‚òëÔ∏è Visualization of Training Sets Results

```python
plt.scatter(x_train, y_train, color = 'green')
plt.plot(x_train, regressor.predict(x_train)) # Adjust. line plotted with '''x_train & predict(x_train)'''
plt.legend()
plt.xlabel('Years of Exp.')
plt.ylabel('Salary Level')
plt.show()
```

#### ‚òëÔ∏è Visualization of Testing Sets Results

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'darkgrey') # Adjust. line again !!!
plt.xlabel('Years of Exp.')
plt.ylabel('Salary Level')
plt.show()
```

#### ‚òëÔ∏è Predictions of the salary of an employee with 20 years of experience

```python
y_20 = regressor.predict([[20]])
print('His salary is {}'.format(y_20))
```

### üëâüèæüëâüèæ Final Linear Equation 

```python
print(regressor.coef_)
print(regressor.intercept_)
```
