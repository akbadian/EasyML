### üëâüèæ Data Preprocessing

#### ‚òëÔ∏è Importing libraries and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/HP Elitebook/OneDrive/Bureau/HOMEWORKS/ML_BOOTCAMP/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv')
data
```

#### ‚òëÔ∏è Creating and Reshaping Variables

```python
x = data.iloc[:, 1].values
y = data.iloc[:, 2].values
x = x[:, np.newaxis]
```

#### ‚òëÔ∏è Particularity

This time there is no missing or invalid data so we skip imputation part.<br> Also we've got a small dataset so no need to split the dataset. We will train our model on the wole dataset as we need maximum information for carrying out the model.

### üëâüèæ Training on the whole dataset

#### ‚òëÔ∏è Linear Regression

```python
from sklearn.linear_model import LinearRegression
l_reg = LinearRegression()
l_reg.fit(x,y)
```

#### ‚òëÔ∏è Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
p_reg = PolynomialFeatures( degree = 2)    # Creating Polyn.features with 2 degrees.
x_poly = p_reg.fit_transform(x)            # Transforming the independant variable 'x' into a polynomial features
l_reg2 = LinearRegression()
l_reg2.fit(x_poly, y)                      # Fitting a new linear_reg model with our polynomial ind_var 'x_poly'
```

### üëâüèæ Visualization

```python
plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
plt.scatter(x, y, c = 'red')
plt.plot(x, l_reg.predict(x), c = 'blue')
plt.title('Linear Regression Version')
plt.xlabel('Salary level')
plt.ylabel('Salaries')

plt.subplot(1,2,2)
plt.scatter(x, y, c = 'red')
plt.plot(x, l_reg2.predict(x_poly), c = 'blue')
plt.title('Polynomial Regression Version')
plt.xlabel('Salary level')
plt.ylabel('Salaries')
```

### üëâüèæ Training Polynomial Regression with Higher Degree For much Accuracy

#### ‚òëÔ∏è Fitting the model

```python
p_reg2 = PolynomialFeatures( degree = 4)    # Creating Polyn.features with 4 degrees.
x_poly2 = p_reg2.fit_transform(x)            # Transforming the independant variable 'x' into a polynomial features
l_reg3 = LinearRegression()
l_reg3.fit(x_poly2, y)                      # Fitting a new linear_reg model with our polynomial ind_var 'x_poly'
```

#### ‚òëÔ∏è Visualization

```python
plt.figure(figsize = (12,6))

plt.subplot(1,2,1)
plt.scatter(x, y, c = 'red')
plt.plot(x, l_reg2.predict(x_poly), c = 'blue')
plt.title('Polynomial Regression Version - 2 degrees')
plt.xlabel('Salary level')
plt.ylabel('Salaries')

plt.subplot(1,2,2)
plt.scatter(x, y, c = 'red')
plt.plot(x, l_reg3.predict(x_poly2), c = 'blue')
plt.title('Polynomial Regression Version - 4 degrees')
plt.xlabel('Salary level')
plt.ylabel('Salaries')
```
